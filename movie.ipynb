```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
import ast#for converting dictionary to list
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert1(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break    
    return L    
```

```python
movies['cast']=movies['cast'].apply(convert1)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Removing spaces to avoid confusion during tags
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags']=movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movie=movies[['title','id','tags']]
```

TEXT VECTORIZATION

```python
#converting tags to vectors

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

 
```

```python
vectors=cv.fit_transform(new_movie['tags']).toarray()
```

```python
#applying stemming
#['loved,loving,love]->[love,love,love]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python



def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
```

```python
new_movie['tags'][0]
```

```python
new_movie['tags']=new_movie['tags'].apply(stem)  #STEMMING APPLIED
```

```python
#Calculating distance between vectors using cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

A function to  recommend five  similar movies

```python
list(enumerate(similarity[0]))#holds the index position of the cosine distancces of  each movie
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]#sorting is done on second number basis
```

Recommend 

```python
def recommend(movie):
    movie_index=new_movie[new_movie['title']==movie].index[0] 
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i  in movies_list:
        print(new_movie.iloc[i[0]].title)
        
        
```

```python
recommend('Batman Begins')
```
